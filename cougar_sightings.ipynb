Plot the Frequency of the Sighting Classes with a Bar Graph

```python
value_counts = data['SightingClass'].value_counts()

# Plot the bar graph
value_counts.plot(kind='bar')
plt.xlabel('Attribute Values')
plt.ylabel('Frequency')
plt.title('Frequency of Attribute Values')
plt.show()
```

```python
# Plot the pie chart
fig, ax = plt.subplots(figsize=(10, 6))
wedges, texts, autotexts = ax.pie(value_counts, startangle=90, autopct='')

# Remove the ylabel to make it cleaner
ax.set_ylabel('')

# Set the title
ax.set_title('Frequency of Attribute Values')
ax.axis('equal')

# Combine the labels with their corresponding percentages
total = value_counts.sum()
percentage_labels = [f'{label}: {count / total * 100:.1f}%' for label, count in zip(value_counts.index, value_counts)]

# Add a legend to the side with the percentage labels
ax.legend(wedges, percentage_labels, loc="best", bbox_to_anchor=(1, 0, 0.5, 1))

# Show the plot
plt.show()
```

FEATURE CLASSES FOR SIGHTINGS

Created FC for Class 1a 

```python
# Set Excel File and set Sighting Class for Query 
csv_file = csv_file = 'https://raw.githubusercontent.com/ecorey/Eastern_Cougar_Arcpy_/main/sightingTable_with_degrees.csv'

data = pd.read_csv(csv_file)

filtered_data = data[data['SightingClass'] == 'Class 1a']

# NEED TO SET FOT LOCAL SYSTEM
# Set a parent folder for the new file geodatabase
parent_folder = r"C:\Users\xyz\arcGIS_test\test"

# Set the name of the new file geodatabase
gdb_name = "cougar_sighting_points.gdb"

# Create the new file geodatabase
arcpy.CreateFileGDB_management(parent_folder, gdb_name)

# NEED TO SET FOT LOCAL SYSTEM
# Set the output workspace
arcpy.env.workspace = r'C:\Users\xyz\arcGIS_test\test\cougar_sighting_points.gdb'
arcpy.env.overwriteOutput = True

# set the spatial reference system
sr = arcpy.SpatialReference(4269)  # WGS 84

# NEED TO SET FOT LOCAL SYSTEM
# output path
outFCPt = r'C:\Users\xyz\arcGIS_test\test\cougar_sighting_points.gdb\fc_pt_1a'
geometry_type = "POINT"

arcpy.CreateFeatureclass_management(os.path.dirname(outFCPt),
                                    os.path.basename(outFCPt),
                                    "Point", spatial_reference=sr)

# Fields in the Output Feature Class 
fields = arcpy.ListFields(outFCPt)
field_names = [f.name for f in arcpy.ListFields(outFCPt)]
print(field_names)

# Creating Point Features from Extracted Coordinates 
parray = arcpy.Array()  # An array to hold point objects
point = arcpy.Point()  # Initialize a point object

# Iterate over the filtered_data DataFrame
for _, row in filtered_data.iterrows():
    lat = row['Y_Cord']  
    lon = row['X_Cord']  

    point.X = lon
    point.Y = lat
    parray.append(point)

# Creating Point Features from Extracted Coordinates 
with arcpy.da.InsertCursor(outFCPt, ('SHAPE@XY')) as icur2:  
    for p in parray:
        icur2.insertRow([p])

# Retreive Values From FC for POINT Values        
fields = ['SHAPE@XY']

with arcpy.da.SearchCursor(outFCPt, fields) as cursor:
    for row in cursor:
        print(f'{row[0][0]:.2f}')
        print(f'{row[0][1]:.2f}')

# List Feature Class
featureclasses = arcpy.ListFeatureClasses()

for fc in featureclasses:
    print(fc)

# Print number of points
num_points = len(parray)
print(f"Total number of points created: {num_points}")
```
